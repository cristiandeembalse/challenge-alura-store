# **Introducci√≥n**

### **Estructura de datos**
#### El conjunto de datos incluye informaci√≥n sobre:

üè∑ **Categor√≠a**: Clasificaci√≥n del producto.

### **An√°lisis realizados**
1. An√°lisis de facturaci√≥n.

# **6. Generando los gr√°ficos**

# **7. An√°lisis del desempe√±o geogr√°fico**

```python
import seaborn as sns
```

```python
tabla_tienda1 = tienda.groupby(['lat', 'lon']).size().unstack(fill_value=0)
tabla_tienda2 = tienda2.groupby(['lat', 'lon']).size().unstack(fill_value=0)
tabla_tienda3 = tienda3.groupby(['lat', 'lon']).size().unstack(fill_value=0)
tabla_tienda4 = tienda4.groupby(['lat', 'lon']).size().unstack(fill_value=0)
```

```python
# Crea una figura con 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Heatmap para la Tienda 1
sns.heatmap(tabla_tienda1, annot=True, cmap='YlGnBu', ax=axes[0, 0])
axes[0, 0].set_title('Tienda 1')

# Heatmap para la Tienda 2
sns.heatmap(tabla_tienda2, annot=True, cmap='YlGnBu', ax=axes[0, 1])
axes[0, 1].set_title('Tienda 2')

# Heatmap para la Tienda 3
sns.heatmap(tabla_tienda3, annot=True, cmap='YlGnBu', ax=axes[1, 0])
axes[1, 0].set_title('Tienda 3')

# Heatmap para la Tienda 4
sns.heatmap(tabla_tienda4, annot=True, cmap='YlGnBu', ax=axes[1, 1])
axes[1, 1].set_title('Tienda 4')

plt.tight_layout()  # Ajusta los subplots para que no se superpongan
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(anos, facturacion_anual, label='Tienda 1', color='royalblue')
plt.plot(anos, facturacion2_anual, label='Tienda 2', color='green')
plt.plot(anos, facturacion3_anual, label='Tienda 3', color='brown')
plt.plot(anos, facturacion4_anual, label='Tienda 4', color='tomato')
plt.xticks([2020, 2021, 2022])
plt.title('Evoluci√≥n de la facturaci√≥n anual de cada tienda')
plt.xlabel('A√±os')
plt.ylabel('Facturaci√≥n')
plt.legend()
plt.show()
```
